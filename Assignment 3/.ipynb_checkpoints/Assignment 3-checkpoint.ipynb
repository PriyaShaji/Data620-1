### Objective and Dataset

In this assignment we'll use a simple 2-node network from Davis Southern Club Women dataset, The data represent observed attendance at 14 social events by 18 Southern women. The result is a person-by-event matrix: cell (i,j) is 1 if person i attended social event j, and 0 otherwise. 

```python
import matplotlib.pyplot as plt
import networkx as nx
from IPython.display import YouTubeVideo
# Import the bipartite Module :  Bipartite graphs B = (U, V, E) have two node sets U,V and edges in 
# E that only connect nodes from opposite sets. 
import networkx.algorithms.bipartite as bipartite 
```

```python
G = nx.davis_southern_women_graph() # Returns Davis Southern women social network.

# Separates the two sets of nodes into 'women' and the other 'clubs'
women = G.graph['top']
clubs = G.graph['bottom']

for line in nx.generate_adjlist(G):
    print(line)
```

```python
# Draw the bipartite graph of women and clubs
plt.figure(3,figsize=(12,12))
nx.draw(G, pos=nx.fruchterman_reingold_layout(G), node_color='#A0CBE2', width=.3, with_labels=True, node_size = 100)
plt.show()
```

#### Clustering Coefficient

This measure will help us understand how many clusters exists between the women and between the events.


```python
import pandas as pd

Events = bipartite.weighted_projected_graph(G, clubs) # Returns a weighted projection of G onto one of its node sets.
Women  = bipartite.weighted_projected_graph(G, women)

# Compute the clustering coefficient for nodes.
Women_cluster  = pd.DataFrame(zip(nx.clustering(Women).keys(), nx.clustering(Women).values())).sort_values(by = [1], axis=0, ascending=True)
# Display Women Cluster
Women_cluster
```

```python
Events_cluster = pd.DataFrame(zip(nx.clustering(Events).keys(), nx.clustering(Events).values())).sort_values(by = [1], axis=0, ascending=True)
# Display Event Cluster
Events_cluster
```

### Cliques

We use the Clicque() functions to find and manipulte cliques. from the graph show below we can see that there are 3 cliques forming within the group of women. We see that Flora Price, Olivia Carleton and Charlotte McDowd are part of one clique only whereas many of the other women are part of the other cliques. We were thinking that maybe the other women are attending several social events to gain status and the more influencial women only have to attend a few as they have reached a higher status within the group.


```python
Women = bipartite.weighted_projected_graph(G, women)
Wcliques = nx.make_clique_bipartite(Women, G)

plt.figure(3,figsize=(12,12))
nx.draw(Wcliques, pos=nx.fruchterman_reingold_layout(Wcliques), node_color='y', alpha = .7, edge_color='red', width=.3, with_labels=True, node_size = 200)
plt.show()
```

It appears that there are two cliques from the events which we expected from our clustering coefficients. The events with .67 for the clustering coefficient are the connections between the two cliques. We would be curious to find out what kind of social events these truly are as it seems that few attend both groups of events. If the importance of the events or types were available I suspect that a pattern may be more easily visible.

```python
Events = bipartite.weighted_projected_graph(G, clubs)
E_cliques = nx.make_clique_bipartite(Events, G)
plt.figure(3,figsize=(12,12))
nx.draw(E_cliques, pos=nx.fruchterman_reingold_layout(E_cliques), node_color='y', alpha = .8, edge_color='red', width=.3, with_labels=True, node_size = 100)
plt.show()
```

### Ego graph

Based on the clustering coefficient we looked at Flora Price and Brenda Rogers, as we can see Flora price has fewer connections and only connected to one cluster. Brenda Rogers went to many social events and has a lower cluster coefficient as she is connected to several clusters.

```python
x = nx.ego_graph(G, 'Flora Price', center = True, undirected = True)
print(nx.info(x))
plt.figure(3,figsize=(7,7))
nx.draw(x,  node_color='y', alpha = .7, edge_color='red', width=.8, with_labels=True, node_size = 100)
```

```python
x = nx.ego_graph(G, 'Brenda Rogers', center = True, undirected = True)
print(nx.info(x))
plt.figure(3,figsize=(7,7))
nx.draw(x,  node_color='y', alpha = .8, edge_color='red', width=.8, with_labels=True, node_size = 100)
```
