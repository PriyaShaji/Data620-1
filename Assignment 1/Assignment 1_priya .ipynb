In this assignment, we are asked to create nodes and edges for a Krackhardt kite graph.

We will use NetworkX package to create the netwrok.

Import required libraries

```python
import networkx as nx
import matplotlib.pyplot as plt

## inline is used to draw graph inline in notebook

%matplotlib inline
```

A graph has two main components: 

Nodes: individual components of the graph

Edges: Distance and direction between two nodes

Now, we will get information for nodes and edges of Krackhardt kite graph from Krackhardt_kite_graph() function

```python
g = nx.generators.small.krackhardt_kite_graph()
```

Nodes and edges of Krackhardt kite graph 

```python
## optional 

print("Nodes of krackhardt kite graph: ","\n",g.nodes(),"\n")

print("Edges of krackhardt kite graph: ","\n", g.edges())
```

```python
## No. of Edges

g.number_of_edges()
```

```python
## No. of Nodes

g.number_of_nodes()
```

Draw Krackhardt kite graph

```python
nx.draw(g, with_labels=1)
plt.show()
```

Substitute node labels with given names

```python
#node_name ={9:"Jane",8:"Ike",7:"Heather",5:"Fernando",6:"Garth",2:"Carol",4:"Ed",3:"Diane",0:"Andre",1:"Beverly"}

node_name = {2:"Carol",5:"Fernando",7:"Heather",8:"Ike",9:"Jane",6:"Garth",4:"Ed",1:"Beverly",3:"Diane"
         ,0:"Andre"}
node_name
```

```python
# create Graph object
G =nx.Graph()

# create Graph Edge object


e = g.edges()

# Add all the Edges to the Graph object 

G.add_edges_from(e)

print("Edges for Named network:", "\n" ,G.edges())
```

```python
# Relabel nodes with node_names

H = nx.relabel_nodes(G,node_name)
nx.draw(H, with_labels=1)
plt.show()
```
