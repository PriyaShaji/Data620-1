### Objective

In this notebook, we will work with a social network of seventh graders, in which nodes are individual students, and edges represent their relationships. Edges between individuals show how often the seventh graders indicated other seventh graders as their favourite. We'll create basic analysis on the graph, including the graphâ€™s diameter, 

### Dataset

This [dataset](http://konect.uni-koblenz.de/networks/moreno_seventh) comes from a study of 7th grade students.
This directed network contains proximity ratings between studetns from 29 seventh grade students from a school in Victoria. Among other questions the students were asked to nominate their preferred classmates for three different activities. A node represents a student. An edge between two nodes shows that the left student picked the right student as his answer. The edge weights are between 1 and 3 and show how often the left student chose the right student as his favourite.

```python
# Import Modules 

import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

#### Read the dataset and create the Network elements

```python
#read the edge list
edges = pd.read_csv('moreno_seventh/seventh_seventh', skiprows=2, sep = " ", header=None)
edges.columns = ['student1', 'student2', 'count']

#read node metadata
nodes =pd.read_csv('moreno_seventh/student_gender', header=None)
nodes.index += 1
nodes.columns = ['gender']

#construct the graph
G = nx.DiGraph()
for row in edges.iterrows():
    G.add_edge(row[1]['student1'], row[1]['student2'], count=row[1]['count'])


#add meta data
for i in G.nodes():
    G.nodes[i]['gender'] = nodes.loc[i]['gender']
```

**Let's Check the number of nodes**

```python
len(G)
```

**Lets Chech the number of edges**

```python
G.number_of_edges()
```

**Let's get a list of nodes with their attributes.**

```python
list(G.nodes(data=True))[0:5]
```

**Let's Graph the Network using Networkx draw() method**

```python
nx.draw(G, with_labels=True)
```

#### Metrics

**Diameter**

There are many network metrics derived from shortest path lengths. One such measure is diameter, which is the longest of all shortest paths. After calculating all shortest paths between every possible pair of nodes in the network, diameter is the length of the path between the two nodes that are furthest apart. The measure is designed to give you a sense of the networkâ€™s overall size, the distance from one end of the network to another.

```python
nx.diameter(G)
```

**Density** 

Another metric is network density. This is simply the ratio of actual edges in the network to all possible edges in the network. 

```python
density = nx.density(G)
print("Network density:", density)
```

In this case, the density of our network is approximately 0.463. On a scale of 0 to 1, not a very dense network, which comports with what you can see in the visualization

**Centrality**

Degree is the simplest and the most common way of finding important nodes. As we can notice that the degree centrality of node 9 is the maximum. 

```python
nx.degree_centrality(G)
```

```python
# The neighbors()method returns the neighbors of that particular node.

list(G.neighbors(9))
```

### Graphing with Neo4j

To visualize the Netwrorkx Graph in Neo4j, we write the Graph to GraphML and import it into Neo4j. 
GraphML is used by other tools, like Gephi and CytoScape, to read graph data

```python
# Write graphml file
nx.write_graphml(G, 'students.graphml')
```
